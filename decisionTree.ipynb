```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
my_data=pd.read_csv("drug200.csv",delimiter=",")
my_data[0:5]
```

## pre-processing

```python
X=my_data[['Age','Sex',	'BP',	'Cholesterol',	'Na_to_K']].values
X[0:5]
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
le_sex.fit(['F','M'])
X[:,1]=le_sex.transform(X[:,1])


le_BP=LabelEncoder()
le_BP.fit(['LOW','NORMAL','HIGH'])
X[:,2]=le_BP.transform(X[:,2])

le_chol=LabelEncoder()
le_chol.fit(['NORMAL','HIGH'])
X[:,3]=le_chol.transform(X[:,3])

X[0:5]
```

```python
y=my_data['Drug']
y[0:5]
```

## Setting up the Decision Tree

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=3)
```

## Modeling

```python
drugTree=DecisionTreeClassifier(criterion="entropy",max_depth=4) #make decision tree. criterion can be "entrophy" or "gini"
drugTree
```

```python
drugTree.fit(x_train,y_train)
```

## Prediction

```python
predDrug=drugTree.predict(x_test)
```

```python
print(predDrug[0:5])
print(y_test[0:5])
```

## Evaluation

```python
from sklearn import metrics
print("DecisionTrees's Accuracy:", metrics.accuracy_score(y_test,predDrug))
```
